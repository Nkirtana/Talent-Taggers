```python
# For plotting the data
import matplotlib.pyplot as plt


# to read multiple files or a single file from a directory
from filereader import ReadFiles

from datacleaning import CleanData

# NLTK imports
from nltk.tokenize  import word_tokenize, sent_tokenize
```

```python
# Enter dataset path here
files_path = r"C:\Users\manoj\Desktop\dataset"
```

```python
file_reader_obj = ReadFiles(files_path)
file_reader_obj.read_input()
```

```python
resumes = list(file_reader_obj.text_dict.values())
```

```python
resumes_text = list()
for resume in resumes:
    cd = CleanData()
    cd.initialize_text(resume)
    cd.clean_text()
    resumes_text.append(cd.get_output()['output_text'].lower())
    
```

```python
len(resumes_text)
```

```python
combined_resume_text = " ".join(resumes_text)
```

```python
toks = word_tokenize(combined_resume_text)
```

```python
from nltk import FreqDist
```

```python
words = list()
counts = list()
for word, freq in FreqDist(toks).most_common(40):
    if len(word)==1:
        continue
    words.append(word)
    counts.append(freq)
```

```python
font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }


plt.figure(figsize=(15,5))
plt.bar(words, counts, align='center', width= 0.7)
plt.xlabel("Word",fontdict=font )
plt.xticks(rotation=60, )
plt.ylabel("word count", fontdict=font, fontsize = 10, va='bottom', ha='left')
```
